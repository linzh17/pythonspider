# requests 库的使用

## resquests.get()
> 与urllib中urlopen()方法类似 

```python
import requests

r = requests.get('http://www.baidu.com/')
print(type(r))
print(r.status_code)
print(type(r.text))
print(r.text)
print(r.cookies)
```

```python
import requests

#添加参数,构建字典传给params
data = {
    'name':'germey',
    'age':22
}
r = requests.get('http://httpbin.org/get',params=data)
print(r.text)
```

> 上述例子网页返回结果为str类型,是json格式,如果想直接解析返回结果,得到一个字典格式,可以调用json()  
> 如果返回的不是json格式,便会出现解析错误,抛出json.decoder.JSONDecodeError异常


```python
import requests

r = requests.get('http://httpbin.org/get')
print(type(r))
print(r.json())
print(type(r.json()))
```

```python
#抓取网页

import requests
import re

headers  = {
    'user-agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/69.0.3497.100 Safari/537.36'
}

r = requests.get('https://www.zhihu.com/explore',headers=headers)
pattern  = re.compile('explore-feed.*?question_link.*?>(.*?)</a>',re.S)
titles = re.findall(pattern,r.text)
print(titles)
```

> 抓取二进制数据 上述例子抓取的是网页 ,返回的时html文档,而抓取图片,音频,视频等二进制文件  


```python
import requests 

r = requests.get('https://github.com/favicon.ico')
#print(r.text) #会出现乱码 原因 图片直接转化为字符串
#print(r.content)

with open('favicon.ico','wb') as f:
    f.write(r.content)
```

> 有的网站,不添加headers,不能正常请求,需要添加headers 例如知乎

```python
import requests

r= requests.get('https://www.zhihu.com/explore')#获取网页内容失败
print(r.text)

headers  = {
    'user-agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/69.0.3497.100 Safari/537.36'
}

r= requests.get('https://www.zhihu.com/explore',headers=headers)#添加headers 获取成功
print(r.text)

```

## 响应
> 响应的几个属性

```python
import requests

r = requests.get('https://httpbin.org/get')
if not r.status_code ==requests.codes.ok:
    exit() 
else:
    print('Requests Ok')
```

## 会话维持
> 维持同一个会话 使用同一个session 对象

```python
import requests 

s = requests.Session()
s.get('http://httpbin.org/cookies/set/number/123456789')
r = s.get('http://httpbin.org/cookies')
print(r.text)
```

## SSL 证书验证
> requests 提供证书验证的功能,由verify参数控制是否检查证书,默认为True 自动验证


```python
# example verify = True 

import requests 

response =  requests.get('https://www.12306.cn')
print(response.status_code)
```

```python
import requests

#设置为False,输出结果 会有警告 提示指定证书
response = requests.get('https://www.12306.cn',verify=False)
print(response.status_code)

#指定本地证书用作客户端证书,本地私有证书的key必须是解密状态
#response = requests.get('https://www.12306.cn',cert=('path/server.crt','/path/key'))
```

## 代理设置
> 与urllib 的代理设置类似

```python
import requests 

proxies = {
    'http':'http://10.10.1.10:3128',
    'https':'https://10.10.1.10:3128',

}

response = requests.get('https://www.taobao.com',proxies=proxies)

# 若代理需要使用http basic auth 则 使用 类似http://user:password@host:port

proxies = {
    'http':'http://user:password@10.10.1.10:3128'
}

response = requests.get('https://www.taobao.com',proxies = proxies)

# 支持socks协议的代理

proxies = {
    'http':'socks5://user:password@host:port',
    'https':'socks5://user:password@host:port'
}

response = requests.get('https://www.taobao.com',proxies = proxies)
```

## 超时设置
> 指定timeout 参数  
> timeout 指定连接和读取两部分的总和时间,也可以分别指定,需要用传入一个元组  
> timeout=None  永久等待,不会出现超时错误

```python
import requests 

r = requests.get('https://www.taobao.com',timeout=1)
print(r.status_code)

r = requests.get('https://www.taobao.com',timeout=(5,11))
print(r.status_code)

r = requests.get('https://www.taobao.com',timeout=None)
print(r.status_code)
```

## 身份认证
> 用于需要进行身份验证的页面  
> 常用 HTTPBasicAuth 还有OAuth

```python
import requests 
from requests.auth import HTTPBasicAuth

r = requests.get('http://httpbin.org',auth = HTTPBasicAuth('2','2'))
print(r.status_code)

#可以简写

r = requests.get('http://httpbin.org',auth = ('2','2'))
print(r.status_code)
```

```python
from requests import Request,Session

url = 'http://httpbin.org/post'

data = {
    'name':'germey'
}

headers = {
    'User-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/69.0.3497.100 Safari/537.36',
}

s = Session()

req = Request('POST',url,data=data,headers=headers)
prepped = s.prepare_request(req)
r = s.send(prepped)
print(r.text)
```
